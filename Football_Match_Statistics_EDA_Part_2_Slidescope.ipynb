```python
# Map of incorrect team names to correct ones
team_replacements = {
    "Hoffenheim\n2": "Hoffenheim",
    "Eintracht Frankfurt\n2": "Eintracht Frankfurt",
    "Darmstadt\n2": "Darmstadt",
    "CD Badajoz\r\n2": "CD Badajoz",
    "Hertha Berlin\n2": "Hertha Berlin"
}

# Replace the team names in 'away_team'
df['away_team'] = df['away_team'].replace(team_replacements)
df['home_team'] = df['home_team'].replace(team_replacements)
```

```python
df[df['home_team'].str.contains('\n2')]['home_team']

names_list = list(df[df['home_team'].str.contains('\n2')]['home_team'])
```

```python
names_list
```

```python
# prompt: in names_list there are team names containing '\r\n2' write code to replace these names in home team with their correct name as done above

# Create a dictionary to map incorrect team names to correct ones
# This dictionary should include all the names in names_list
team_replacements_home = {name: name.replace('\n2', '') for name in names_list}

# Replace the team names in 'home_team' using the dictionary
df['away_team'] = df['away_team'].replace(team_replacements_home)
```

```python
# Convert 'home_score' and 'away_score' columns to numeric,
# coercing errors to NaN (Not a Number)
df['home_score'] = pd.to_numeric(df['home_score'], errors='coerce')
df['away_score'] = pd.to_numeric(df['away_score'], errors='coerce')

# Now, aggregate home and away scores
team_scores_home = df.groupby('home_team')['home_score'].sum().reset_index()
team_scores_away = df.groupby('away_team')['away_score'].sum().reset_index()
```

```python
# Merge on 'team'
team_scores = pd.merge(team_scores_home, team_scores_away, on='team', how='outer').fillna(0)
```

```python
# Aggregate home and away scores
team_scores_home = df.groupby('home_team')['home_score'].sum().reset_index()
team_scores_away = df.groupby('away_team')['away_score'].sum().reset_index()
```

```python
# Aggregate expected goals for home and away teams
team_scores_home_expected = df.groupby('home_team')['expected_goals_xg_home'].sum().reset_index()
team_scores_away_expected = df.groupby('away_team')['expected_goals_xg_host'].sum().reset_index()
```

```python
# Total Goals Conceded by teams in (Home vs Away)
```

```python
# Aggregate home and away scores
team_scores_home = df.groupby('home_team')['away_score'].sum().reset_index()
team_scores_away = df.groupby('away_team')['home_score'].sum().reset_index()

# Rename columns for consistency
team_scores_home.rename(columns={'home_team': 'team', 'away_score': 'home_conceded'}, inplace=True)
team_scores_away.rename(columns={'away_team': 'team', 'home_score': 'away_conceded'}, inplace=True)

# Merge on 'team'
team_scores = pd.merge(team_scores_home, team_scores_away, on='team', how='outer').fillna(0)

# Add a column for the total goals scored by each team (home + away)
team_scores['Total_Score'] = team_scores['home_conceded'] + team_scores['away_conceded']

# Sort by total goals scored (Total_Score)
team_scores = team_scores.sort_values(by='Total_Score', ascending=True)

# Reshape the data for plotting
team_scores_melted = team_scores.melt(id_vars='team',
                                      value_vars=['home_conceded', 'away_conceded'],
                                      var_name='Concede_Type',
                                      value_name='Concede')

```

```python
team_scores_melted
```

```python

# Plot the data
plt.figure(figsize=(20, 13))
sns.barplot(data=team_scores_melted[team_scores_melted['team'].isin(['Barcelona', 'Real Madrid', 'Juventus', 'Manchester City', 'Arsenal', 'Chelsea'])], x='team', y='Concede', hue='Concede_Type', palette='coolwarm', ci=None)

# Customizing the plot
plt.xticks(rotation=45, ha='right', fontsize=20)  # Rotate labels for better readability
plt.xlabel('Team', fontsize=20)
plt.ylabel('Total Goals Conceded', fontsize=20)
plt.title('Total Goals Conceded in  (Home vs Away)', fontsize=20)
plt.legend(title='Concede Type', fontsize=25)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for better visibility
plt.tight_layout()
plt.show()
```

1) Most teams tended to score more goals during their home matches compared to their away games, while conceding more goals when playing away. <br/>  
2) Teams at the top of the points table outperformed their expected goals (xG), scoring more than anticipated, whereas teams in the bottom half fell short, scoring fewer goals than expected.
